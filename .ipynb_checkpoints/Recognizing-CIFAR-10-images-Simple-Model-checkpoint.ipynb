## Recognizing CIFAR-10 images with Deep Learning (Part I)

CIFAR-10 dataset contains 60,000 color images of size 32*32 pixels which are divided in 10 classes.
We will use Keras library to import the dataset which contains 50,000 images as training iamges
and 10,000 as test images

```python
import numpy as np
from keras.datasets import cifar10
from keras.utils import np_utils
import operator
#from keras.optimizers import SGD, Adam, RMSprop
#from keras.models import Sequential
#from keras.layers.convolutional import Conv2D, MaxPooling2D
#from keras.layers.core import Dense, Flatten, Activation, Dropout
import matplotlib.pyplot as plt
% matplotlib inline
```

```python
#load dataset
(X_train, y_train), (X_test, y_test) = cifar10.load_data()
```

```python
BATCH_SIZE = 128
nb_epochs = 20
VALIDATION_SPLIT = 0.2
num_classes = 10
```

```python
num_train, img_channels, img_rows, img_cols = X_train.shape
num_test, _, _, _= X_test.shape
```

```python
print('X_train Shape:', X_train.shape)
training_samples = X_train.shape[0]
print("Training samples:", training_samples)
print("Test Samples:", y_test.shape[0])
```

```python
def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.2989, 0.5870, 0.1140]).flatten()
```

```python
X_train_gray = []
for idx, x in enumerate(X_train):
    X_train_gray.append(rgb2gray(x))
X_train_gray = np.array(X_train_gray) 
```

```python
X_test_gray = []
for idx, x in enumerate(X_test):
    X_test_gray.append(rgb2gray(x))
X_test_gray = np.array(X_test_gray) 
X_test_gray = X_test_gray.astype('float32')
X_test_gray /= 255
X_test_gray = X_test_gray.tolist()
```

```python
class_names =['airplane', 'automobile', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck']
```

One hot Encoding

```python
fig = plt.figure(figsize =(8, 3))
for i in range(num_classes):
    axis = fig.add_subplot(2, 5, 1+i, xticks = [], yticks = [])
    idx = np.where(y_train[:] == i)[0]
    features_idx = X_train_gray[idx]
    image_num = np.random.randint(features_idx.shape[0])
    image = features_idx[image_num]
    axis.set_title(class_names[i])
    plt.imshow(image.reshape((32,32)))
plt.show()
```

```python
X_train_gray = X_train_gray.astype('float32')
X_train_gray /= 255
X_train_gray = X_train_gray.tolist()
```

```python
# convert to categorical
y_train = np_utils.to_categorical(y_train, num_classes)
y_test = np_utils.to_categorical(y_test, num_classes)
```

```python
from Neural_Network import Neural_Network
from Neural_Network import Activation_Function
from Neural_Network import Loss

nn = Neural_Network()
nn.add_input_layer([1024])
nn.add_flatten_layer()
nn.add_fully_connected_layer(750, Activation_Function.Relu)
nn.add_dropout_layer(0.1)
nn.add_fully_connected_layer(500, Activation_Function.Relu)
nn.add_dropout_layer(0.1)
nn.add_fully_connected_layer(250, Activation_Function.Relu)
nn.add_dropout_layer(0.1)
nn.add_fully_connected_layer(125, Activation_Function.Relu)
nn.add_dropout_layer(0.1)
nn.add_fully_connected_layer(75, Activation_Function.Relu)
nn.add_fully_connected_layer(50, Activation_Function.Relu)
nn.add_fully_connected_layer(25, Activation_Function.Relu)
#nn.add_dropout_layer(0.1)
nn.add_output_layer(10, Activation_Function.SoftMax)
```

```python
nn.build()
nn.print_loss_every_iterations(1)
```

```python
nn.setLearningRate(0.001)
nn.shuffle_training_data(True)
nn.train(X_train_gray, y_train, 64, 10, Loss.CrossEntropy)
```

```python
index = 0
matches = 0
for data in X_train_gray:
    output = nn.execute(X_train_gray[index])
    outputIndex, value = max(enumerate(output), key=operator.itemgetter(1))
    labelIndex, label_value = max(enumerate(y_train[index]), key=operator.itemgetter(1))
    if outputIndex == labelIndex:
        matches += 1
    index += 1
match_percentage = matches / index
print("matches: ")
print(matches)
print("index: ")
print(index)
print("percent correct: ")
print(match_percentage)
```

```python
index = 0
matches = 0
for data in X_test_gray:
    output = nn.execute(X_test_gray[index])
    outputIndex, value = max(enumerate(output), key=operator.itemgetter(1))
    labelIndex, label_value = max(enumerate(y_test[index]), key=operator.itemgetter(1))
    if outputIndex == labelIndex:
        matches += 1
    index += 1
match_percentage = matches / index
print("matches: ")
print(matches)
print("index: ")
print(index)
print("percent correct: ")
print(match_percentage)
```

```python
from keras import backend as K
if K.backend() == 'tensorflow':
    K.set_image_dim_ordering('tf')
```

Building sequential model

```python
# Network
model = Sequential()
model.add(Conv2D(32, (3, 3), padding = 'same', input_shape=X_train.shape[1:]))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size= (2,2)))
model.add(Dropout(0.25))
```

```python
model.add(Flatten())
model.add(Dense(512))
model.add(Activation('relu'))
model.add(Dropout(0.5))
model.add(Dense(num_classes))
model.add(Activation('softmax'))
model.summary()
```

Compiling the model

```python
model.compile(loss= 'categorical_crossentropy', optimizer = 'rmsprop', metrics= ['accuracy'])
```

Training the model

```python
model_history = model.fit(X_train, y_train, batch_size = BATCH_SIZE, epochs = nb_epochs, 
          validation_split = VALIDATION_SPLIT, verbose = 1)
```

List all data in history

```python
print(model_history.history.keys())
```

Summarize history for accuracy

```python
plt.plot(model_history.history['acc'])
plt.plot(model_history.history['val_acc'])
plt.title("Model Accuracy")
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend(['train','validation'], loc = 'upper left')
plt.show()
```

Summarize history for loss

```python
plt.plot(model_history.history['loss'])
plt.plot(model_history.history['val_loss'])
plt.title("Model loss")
plt.xlabel('Epoch')
plt.ylabel('loss')
plt.legend(['train','validation'], loc = 'upper right')
plt.show()
```

```python
score = model.evaluate(X_test, y_test, batch_size = BATCH_SIZE, verbose =1)
```
